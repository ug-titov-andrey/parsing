```python
from lxml import html
import requests
```

```python
USER_AGENT = 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36'
URL_LENTA = 'https://lenta.ru/'
```

```python
headers={'User-Agent':USER_AGENT,}
params  = {'sso_failed': '',}
response = requests.get(URL_LENTA, headers = headers, params = None)
url_lenta= response.url #url-страницы
dom = html.fromstring(response.text) #получилы объектное представление страницы

```

```python
all_news= dom.xpath("//a[contains(@class,'card-')]") # возвращает список всех новостей
list_news = []
for news in all_news: # Перебираем в цикле все новости
    source ='lenta.ru' #lenta сама является источником новостей
    n_text = news.xpath(".//span[contains(@class, 'card-')]/text()")# если возвращается пустой список, то None
    if len(n_text) == 1 :
        n_text = n_text[0]
    else:
        n_text = None  
    n_link = url_lenta + news.xpath("./@href")[0]
        
    n_date = news.xpath(".//time/text()")
    if len(n_date) == 1:
        n_date = n_date[0]
    else:
        n_date = None
    news_dict = {
        'new_source': source,
        'new_text': n_text,
        'new_link': n_link,
        'new_date': n_date,
    }
    list_news.append(news_dict)
        
```

Получили список словарей, содержащих источник (лента), текст новости, ссылку и дату (если есть)

```python
for ind, news in enumerate(list_news):
    print(f'news => {ind}, new_source =>  {source}, new_text =>  {n_text}, new_link => {n_link},\
    new_link => {n_link},new_date => {n_date}' )
```
